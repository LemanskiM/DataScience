```python
def BulidShortcut(row):
    pob = row["Industry"]
    result = ''
    for i in pob.split(' '):
        result += i[0]
    return result
```

```python
BulidShortcut({'Industry':'Factory Underbewspaper'})
```

```python
fortune['IndustryShort'] = fortune.apply(BulidShortcut,axis=1)
```
